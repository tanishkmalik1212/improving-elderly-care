```python
package com.example.elderlycareapp;

import android.os.Bundle;
import android.widget.*;
import androidx.appcompat.app.AppCompatActivity;

public class HealthTrackerActivity extends AppCompatActivity {
    EditText editBP, editSugar;
    Button btnSave;
    TextView txtResult;

    @Override
    protected void onCreate(Bundle savedInstanceState) {
        super.onCreate(savedInstanceState);
        setContentView(R.layout.activity_health_tracker);

        editBP = findViewById(R.id.editBP);
        editSugar = findViewById(R.id.editSugar);
        btnSave = findViewById(R.id.btnSaveHealth);
        txtResult = findViewById(R.id.txtResult);

        btnSave.setOnClickListener(v -> {
            String bp = editBP.getText().toString();
            String sugar = editSugar.getText().toString();
            txtResult.setText("Saved: BP = " + bp + ", Sugar = " + sugar + " mg/dL");
});
}
}
```

```python
package com.example.elderlycareapp;

import android.os.Bundle;
import android.widget.*;
import androidx.appcompat.app.AppCompatActivity;
import java.util.ArrayList;

public class ReminderActivity extends AppCompatActivity {
    ArrayList<String> reminders = new ArrayList<>();
    ArrayAdapter<String> adapter;
    ListView listView;
    EditText editMed, editTime;
    Button btnAdd;

    @Override
    protected void onCreate(Bundle savedInstanceState) {
        super.onCreate(savedInstanceState);
        setContentView(R.layout.activity_reminder);

        listView = findViewById(R.id.listReminders);
        editMed = findViewById(R.id.editMed);
        editTime = findViewById(R.id.editTime);
        btnAdd = findViewById(R.id.btnAddReminder);

        adapter = new ArrayAdapter<>(this, android.R.layout.simple_list_item_1, reminders);
        listView.setAdapter(adapter);

        btnAdd.setOnClickListener(v -> {
            String reminder = editMed.getText().toString() + " at " + editTime.getText().toString();
            reminders.add(reminder);
            adapter.notifyDataSetChanged();
            editMed.setText("");
            editTime.setText("");
   });
}
}
```
